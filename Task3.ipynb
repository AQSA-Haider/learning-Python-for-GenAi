## Dictionaries

## creating a dictionary named price_dict

```python
price_dict={"fruits": 100, "vegetables": 200, "dairy": 300, "meat": 400, "grains": 700, "frozen": 800}
print("The categories available are:" )
print(price_dict)
```

## Adding a new product

```python
price_dict["noodles"]=200
print(price_dict)
```

## updating the price of a product

```python
price_dict["meat"]= 1000
print(price_dict.get("meat"))
```

## Removing a product from the dict

using non-existing key with get() method returns None

```python
deleted_val=price_dict.pop("fashion","product not found")
deleted=price_dict.pop("frozen","product not found")
print(deleted_val)
print(deleted)
print(price_dict)
```
