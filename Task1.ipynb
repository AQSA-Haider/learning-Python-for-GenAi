## Converting Tuple sample_product into list

```python
new=[i for i in sample_product]
```

```python
print(new)
```

```python
print(len(new))
```

```python
for i in enumerate(new):
    print(i)
```

## updating the tuple cells

```python
update=[[i,j,k] for i,j,k in new]
print(update)
```
