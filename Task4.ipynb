## Combined Operation

## firstly creating the product list and price_dict 

```python
product=["handbag","wallet","harry potter","Jewelry","kurta set","watch"]
price_dict={"handbag":1000,"wallet":400,"harry potter":600,"Jewelry":700,"kurta set":500,"watch":800}
categories=["bags", "acessories","novels", "acessories","fashion", "acessories"]
print(product)
print(price_dict)
print(categories)
```

## creating a list of tuples from above list and dict

```python
catalog=[]
for pro, cat in zip(product,categories):
    price=price_dict[pro]
    catalog.append((pro,price,cat))

print("catalog:",catalog)    
```

## creating a dict of lists from catalog

```python
category_to_products={}
for pro, price, cat in catalog:
    if cat not in category_to_products:
        category_to_products[cat]=[]
    category_to_products[cat].append(pro)
print(category_to_products)
```

## category with maximum no. of products

```python
max_category=max(category_to_products,key=lambda k:len(category_to_products[k]))

print(f"category with max products: {max_category}")
print(f"products:{category_to_products[max_category]}")
```
